### Testing con n煤meros reales

Los n煤meros reales son infinitos, pero el computador tiene memoria finita. Luego, no es posible representar directamente todos los n煤meros reales. Por lo tanto, los n煤meros reales se representan con aritm茅tica de punto flotante. 

Luego, inevitablemente al operar con n煤meros reales, nos podemos encontrar con errores de precisi贸n:

```python
>>> 0.1 + 0.1
    0.2
```

```python
>>> 0.1 + 0.2
    0.30000000000000004
```

Por lo que, para testear funciones que operan con n煤meros reales, utilizaremos un **valor de tolerancia o margen de error**.

La idea es indicar que toleraremos un peque帽o error entre el resultado esperado y el resultado entregado por la funci贸n. Para esto, usaremos la funci贸n `cerca`

```python
# cerca: num num num -> bool
# indica si dos cantidades son iguales con 
# precision epsilon
# ej: cerca(0.1, 0.2, 0.1) entrega True
def cerca(x,y,eps):
    return abs(x-y) <= eps
```

Luego, para testear funciones que entregan float, usaremos la funci贸n ``cerca(Rf, Re, eps)``

- ``Rf``:  El resultado que nos entrega la funci贸n
- ``Re``:  El resultado que nosotros esperamos
- ``eps``: La tolerancia o margen de error que aceptaremos

As铆, por ejemplo, para testear una funci贸n que calcula la distancia entre dos puntos, hacemos lo siguiente:

### funci贸n `esPrimo` (versi贸n 1)

Para empezar a trabajar con n煤meros primos, primero necesitamos una funci贸n que nos ayude a detectar si un n煤mero dado es primo o no (lo que se conoce como **primalidad**)

Como primera soluci贸n, planteamos lo siguiente:

- Crear una funci贸n auxiliar que cuente la cantidad de divisores de un n煤mero dado (en el rango `1 a n`)

- Luego en la funci贸n principal, podemos decir que un n煤mero es primo si y solo si tiene exactamente 2 divisores

Para la funci贸n auxiliar (`divisores(n)`) , la idea es hacer un "barrido" en el intervalo $[1,n]$, contabilizando cada vez que un n煤mero sea divisor de $n$. Y podemos usar una variable por omisi贸n, para llevar registro de cual es el n煤mero actual que estamos revisando en tal intervalo

- **Caso recursivo**: Si el n煤mero actual es divisor de , entonces lo contamos y pasamos al siguiente n煤mero del intervalo. Si no, entonces de todas maneras pasamos al siguiente n煤mero del intervalo

- **Caso Base**: Si el n煤mero actual a revisar es , entonces sabemos que es divisor, lo contamos, y terminamos la recursi贸n.


````python
# variante 1: Usar 1 variable por omisi贸n
# para saber en que divisor vamos, y llevar
# la cuenta en la recursi贸n

#divisores: int (int) -> int
#da cuantos divisores tiene un numero >= 1
#ej: divisores(6) entrega 4
def divisores(n, actual=1):
    assert type(n) == int and n>=1

    # Caso Base
    if actual == n:
        return 1

    # Caso Recursivo
    if n%actual == 0:
        return 1 + divisores(n, actual + 1)
    else:
        return divisores(n, actual + 1)

#test
assert divisores(6) == 4
assert divisores(1) == 1
assert divisores(7) == 2
````

````python
# Variante 2: Usar una segunda variable por
# omisi贸n, para llevar expl铆citamente la 
# cuenta de divisores

#divisores: int (int) (int) -> int
#da cuantos divisores tiene un numero >= 1
#ej: divisores(6) entrega 4
def divisores(n, actual=1, cuenta=0):
    assert type(n) == int and n>=1

    # Caso Base
    if actual == n:
        return cuenta + 1

    # Caso Recursivo
    if n%actual == 0:
        return divisores(n, actual+1, cuenta+1)
    else:
        return divisores(n, actual+1, cuenta)

#test
assert divisores(6) == 4
assert divisores(1) == 1
assert divisores(7) == 2
````

Luego, con ayuda de esta funci贸n, finalmente podemos definir la funci贸n `esPrimo`

````python
# esPrimo: int -> bool
# indica si un n煤mero dado es primo o no
# ej: esPrimo(7) entrega True
def esPrimo(n):
    assert type(n) == int and n>=1

    if divisores(n) == 2:
        return True
    else:
        return False

# test
assert esPrimo(7)
assert not esPrimo(4)
assert not esPrimo(1)
````

### funci贸n `esPrimo` (versi贸n 2)

Si bien resolvimos el problema, puede ser costoso de calcular, sobre todo para n煤meros grandes, ya que tenemos que barrer todo un intervalo!

Podemos tomar la idea de las funciones anteriores, juntarlas en una sola, y hacer algunas optimizaciones

Mejora 1: Barrer el intervalo, y terminar si es que nos encontramos con 1 divisor del n煤mero

````python
# esPrimo: int (int) -> bool
# ...
def esPrimo(n, divisor = 2):
    assert type(n) == int and n>=1

    # CB1: 1 no es primo
    if n == 1:          
        return False
    # CB2: 2 es primo
    elif n == 2:        
        return True
    # CB3: Si la revisi贸n lleg贸 hasta n, entonces es primo
    elif n == divisor:  
        return True
    # CB4: Si n es divisible, entonces no es primo     
    elif n%divisor == 0:
        return False
        
    #CR: Verificamos las condiciones anteriores con un nuevo divisor
    return esPrimo(n,divisor+1)
````

Mejora 2: Descartar los n煤meros pares anticipadamente

````python
# esPrimo: int (int) -> bool
# ...
def esPrimo(n, divisor = 3):
    assert type(n) == int and n>=1

    if n == 1:
        return False
    # CB2: Si n es par, solo 2 es primo
    elif n%2 == 0:
        return n == 2
    elif n == divisor:
        return True
    elif n%divisor == 0:
        return False
    # Como descartamos los pares, ahora podemos
    # comprobar divisores saltando de impar en impar
    return esPrimo(n,divisor+2)
````

Mejora 3: Solo es necesario barrer hasta $\sqrt{n}$

````python
# esPrimo: int (int) -> bool
# ...
def esPrimo(n, divisor = 3):
    assert type(n) == int and n>=1

    if n == 1:
        return False
    elif n%2 == 0:
        return n == 2
    # CB3: Si la revisi贸n lleg贸 hasta la ra铆z de n, entonces es primo
    elif n < divisor**2:
        return True
    elif n%divisor == 0:
        return False
    return esPrimo(n,divisor+2)
````

### Funci贸n `primeros`

Creemos la funci贸n `primeros(n)`, que muestra en pantalla los primeros `n` n煤meros primos

Ejemplo: `primeros(5)` muestra en pantalla: 2 3 5 7 11 

Idea:

- Descomponer el problema en instancias mas simples

  - La instancia mas simple es que nos pidan imprimir 0 primos

- Necesitamos saber en que parte del intervalo vamos en la revisi贸n

  - Usamos variables por omisi贸n

![](img1_primeros.svg)


````python
# primeros: int (int) -> None
# imprime en pantalla los primeros n primos
# ej: primeros(4) muestra en pantalla 2 3 5 7
def primeros(n, actual = 2):
    assert type(n) == int and n>=0
    if n == 0:
        return

    if esPrimo(actual):
        print(actual)
        return primeros(n-1, actual+1)

    return primeros(n,actual+1)
````

- Caso Base: Si hay que imprimir 0 primos, entonces no hay que hacer nada

- Caso Recursivo: Si el n煤mero actual es primo, entonces reducimos el problema. Si no, seguimos buscando

### Funci贸n `siguiente`

Creemos la funci贸n `siguiente(n)`, que entrega el primo m谩s cercano a `n` (por la derecha)

Ejemplo: `siguiente(24)` entrega 29

Idea:

- Desde el punto de partida (n), vamos preguntando hacia la derecha hasta encontrar un n煤mero primo

![](img2_siguiente.svg)

````python
# siguiente: int -> int
# entrega el primo mas cercano a un n煤mero dado
# ej: siguiente(24) entrega 29
def siguiente(n):
    assert type(n) == int and n>=0

    if esPrimo(n+1):
        return n+1

    return siguiente(n+1)

# Test
assert siguiente(24) == 29
assert siguiente(7) == 11
assert siguiente(1) == 2
````

- Caso Base: Si el siguiente n煤mero es primo, entonces terminamos

- Caso Recursivo: Seguimos la b煤squeda

### Funci贸n `i-esimo` (variante 1)

Creemos la funci贸n `i_esimo(i)`, que entrega el i-esimo n煤mero primo

Ejemplo: `i_esimo(4)` entrega 7

Idea (variante 1):
- Podemos descomponer el problema, siguiendo la siguiente idea:

  - Buscar al 4掳 primo desde 2, es equivalente a buscar al 3掳 primo desde 3, lo que es equivalente a buscar al 2掳 primo desde 5...

- Podemos usar la funci贸n anterior para ir "saltando" de primo en primo

![](img3_iesimo1.svg)

````python
# i_esimo: int (int) -> int
# entrega el i-esimo primo
# ej: i_esimo(2) entrega 3
def i_esimo(i, p_actual = 2):
    assert type(i) == int and i >=0

    if i == 1:
        return p_actual

    sig = siguiente(p_actual)
    return i_esimo(i-1,sig)

# Test
assert i_esimo(2) == 3
assert i_esimo(3) == 5
assert i_esimo(5) == 11
````

Caso Base: Si preguntan por el 1-esimo primo, entonces entregamos el primo que tenemos guardado

Caso Recursivo: Saltamos al siguiente primo y disminuimos el tama帽o del problema

### Funci贸n `i-esimo` (variante 2)

Una segunda forma de resolverlo, es contar de izquierda a derecha, usando par谩metros por omisi贸n 

Idea (variante 2):

- Necesitamos una forma de ir contando primos desde el principio, hasta eventualmente llegar al n煤mero ``i``

- Podemos usar la funci贸n anterior para ir "saltando" de primo en primo

![](img4_iesimo2.svg)


````python
# i_esimo: int (int) -> int
# entrega el i-esimo primo
# ej: i_esimo(2) entrega 3
def i_esimo(i, cont = 1, p_actual = 2):
    assert type(i) == int and i >=0

    if i == cont:
        return p_actual

    sig = siguiente(p_actual)
    return i_esimo(i, cont+1, sig)

# Test
assert i_esimo(2) == 3
assert i_esimo(3) == 5
assert i_esimo(5) == 11
````

- Caso Base: Si nuestra cuenta de primos lleg贸 al i-esimo, entonces entregamos el primo guardado


- Caso Recursivo: Saltamos al siguiente primo, aumentando el contador
